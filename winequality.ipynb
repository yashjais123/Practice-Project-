# Data Cleaning

```python
data.corr()['quality']
```

```python
data=data.drop(['residual sugar','free sulfur dioxide','pH'],axis=1)
data
```

# * Removing Outliers:
# z-score techinque:

```python
from scipy.stats import zscore
import numpy as np
z_score = zscore(data[['fixed acidity','volatile acidity','chlorides','total sulfur dioxide','density','sulphates','alcohol']])
abs_z_score = np.abs(z_score)
```

```python
filtering_entry=(abs_z_score<3).all(axis=1)
data=data[filtering_entry]
data.describe()
```

```python
plt.figure(figsize=(20,15),facecolor='yellow')
plotnumber=1
for column in data[['fixed acidity','volatile acidity','chlorides','total sulfur dioxide','density','sulphates','alcohol']]:
    if plotnumber<=10:
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```

```python
#Lets create one funntion for convert our target in binary classification

data['best quality'] = [1 if x > 7 else 0 for x in data.quality]

# 1 is for good
# 0 is for bad.
```

```python
data
```

```python
#Independent Column
x=data.drop(columns=['quality','best quality'])
#Dependent Column
y=data['best quality']
```

```python
from sklearn.preprocessing import StandardScaler
```

# check multicollinearity problem find if one feature is dependent on another feature

```python
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=7999)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve,roc_auc_score
```

```python
# model accuracy
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def metric_score(clf,x_train,x_test,y_train,y_test,train=True):
    if train:
        y_pred=clf.predict(x_train)
        
        print("\n===========train Result=============")
        print(f"accuracy score:{accuracy_score(y_train,y_pred)*100:.2f}%")
        
    elif train == False:
        pred=clf.predict(x_test)
        
        print("\n=============test Result=============")
        print(f"accuracy score:{accuracy_score(y_test,pred)*100:.2f}%")
        
        print("\n \n test classification report \n",classification_report(y_test,pred,digits=2))
```

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
metric_score(knn,x_train,x_test,y_train,y_test,train=True)
metric_score(knn,x_train,x_test,y_train,y_test,train=False)
```
