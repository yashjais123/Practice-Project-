# Dividing data in feature and vectors

```python
x=data.drop(columns='CGPA')#Feature
y=data.CGPA#Target
```

```python
#Lets import standardscaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

# Now use VIF for checking multicollinearity problem

```python
#Lets import VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i)for i in range(x_scaled.shape[1])]
vif['feature']=x.columns
```

```python
#Lets check how lookinf VIF score
vif
```

# In VIF we consider threshold 5 VIF Score . if any feature is greater then 5 VIF Score it means it contains multicollinearity problmem/

# So in our dataset no feature have more then 5 vif score . so it means out dataset is not contain any multicollinearity problem.

```python
#lets do train test split
#Import train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=876)
```

# * LinearRegression

```python
#Import Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

# * Ridge Regularization

```python
from sklearn.linear_model import Ridge,RidgeCV
```

```python
ridgecv=RidgeCV(alphas=np.arange(0.001,0.1,0.01),normalize=True)
```

```python
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train,y_train)
```

```python
ridge_model.score(x_train,y_train)
```

```python
ridge_model.score(x_test,y_test)
```

# * Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor()
tree.fit(x_train,y_train)
```

```python
tree.score(x_train,y_train)
```

```python
tree.score(x_test,y_test)
```

# * AdaBoost 

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada=AdaBoostRegressor()
ada.fit(x_train,y_train)
```

```python
y_pred = ada.predict(x_train)
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.r2_score(y_train,y_pred)

print("R Square score",accuracy)
```

```python
# predicting Test data with the model 
y_test_pred = ada.predict(x_test)
```

# * Model Evalution

```python
accuracy = metrics.r2_score(y_test,y_test_pred)

print("R Squaree score:",accuracy)
```

# * Hyperparameter Tuning Using RandomizedSearchCv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params={'n_estimators':[47,50,60,70],'learning_rate':[0.25,0.30,0.40]}

rnd_srch = RandomizedSearchCV(AdaBoostRegressor(),cv=5,param_distributions = params)

rnd_srch.fit(x_train,y_train)

rnd_srch.best_estimator_

ada = AdaBoostRegressor(learning_rate=0.26,n_estimators=70)

ada.fit(x_train,y_train)

y_pred = ada.predict(x_test)

print("*** accuracy post tuning ***")
print(metrics.r2_score(y_test,y_pred))
```
